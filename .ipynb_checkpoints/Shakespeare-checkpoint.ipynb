In this notebook we will attempt to generate Shakespeare-like poems using a collection of Shakespeare's poems as training data.
To accomplish this, we will implement an LSTM using Keras

```python
import numpy as np
from keras.models import Model, load_model
```

```python
train_data = open('shakespeare.txt').read().lower()
alphabet = list(set(train_data))
char_dict = {i:char for i, char in enumerate(sorted(alphabet))}
inverse_char_dict = {char:i for i, char in enumerate(sorted(alphabet))}
seq_length = 40
win_size = 3
X = []
Y = []

for i in range(0, len(train_data) - seq_length, win_size):
    X.append(train_data[i: i + seq_length])
    Y.append(train_data[i + seq_length])
    

num_examples = len(X)

x = np.zeros((num_examples, seq_length, len(alphabet)),dtype=np.bool)
y = np.zeros((num_examples, len(alphabet)), dtype=np.bool)
for i, sent in enumerate(X):
    for t, char in enumerate(sent):
        x[i, t, inverse_char_dict[char]] = 1
    y[i, inverse_char_dict[Y[i]]] = 1
    
m = load_model('model_shakespeare_kiank_350_epoch.h5')
```

```python
m.fit(x, y, batch_size=128, epochs=1)
```

```python
gen_text = ''
    
usr_in = 'Coral is far more red'
sent = ('{0:0>' + str(seq_length) + '}').format(usr_in).lower()
gen_text += usr_in
    
for i in range(400):
    ou = np.zeros((1, seq_length, len(alphabet)))
        
    for t, char in enumerate(sent):
        if char != '0':
            ou[0, t, inverse_char_dict[char]] = 1
    y_predict = m.predict(ou, verbose=0)[0]
    y_predict = np.asarray(y_predict).astype('float64')
    y_predict = np.exp(np.log(y_predict)) / np.sum(np.exp(np.log(y_predict)))
    p = np.random.multinomial(1, y_predict, 1)
    cur_index = np.random.choice(range(len(alphabet)), p = p.ravel())
    cur_char = char_dict[cur_index]
    
    gen_text += cur_char
    sent = sent[1:] + cur_char
    
    if cur_char == '\n':
        continue
```
